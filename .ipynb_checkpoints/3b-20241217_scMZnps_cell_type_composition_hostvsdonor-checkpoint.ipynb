```python
from statsmodels.stats.multitest import multipletests  # Corrects for multiple hypothesis testing
from fisher import pvalue_npy  # Computes Fisher's exact test in vectorized form
import numpy as np  # Numerical operations
import pandas as pd  # Data manipulation and analysis

def OR(adata, cluster_col='Subtissue', cluster='Integument', eps=1e-7, correction='fdr_bh', 
       return_significant=True, p_threshold=0.05, OR_cutoff=None, FE_cutoff=None):
    """
    Calculates Odds Ratio (OR), Fold Enrichment (FE), and significance for gene expression in a cluster.
    
    Parameters:
    - adata: AnnData object containing expression data (sparse matrix format)
    - cluster_col: Column in `adata.obs` denoting cluster assignments
    - cluster: Cluster of interest
    - eps: Small constant to avoid division by zero
    - correction: Multiple hypothesis testing correction method
    - return_significant: If True, filters results by significance
    - p_threshold: Threshold for adjusted p-values
    - OR_cutoff: Minimum OR for filtering results
    - FE_cutoff: Minimum Fold Enrichment for filtering results
    
    Returns:
    - DataFrame with calculated statistics for each gene
    """
    
    # Binarize expression data (convert non-zero to 1)
    detection_data = adata.X.copy()
    detection_data.data[detection_data.data != 0] = 1

    # Binarize cluster membership and make 1D array
    membership = np.array(adata.obs[cluster_col] == cluster)  # Boolean array for cluster membership
    membership_data = np.reshape(membership, (detection_data.shape[0], 1)).astype(int)  # Reshape to match dimensions

    # Calculate contingency table
    tp = np.sum(np.add(membership_data, detection_data.toarray()) == 2, axis=0).astype(np.uint)  # True positives
    fp = np.sum(np.subtract(detection_data.toarray(), membership_data) == 1, axis=0).astype(np.uint)  # False positives
    fn = np.sum(np.subtract(membership_data, detection_data.toarray()) == 1, axis=0).astype(np.uint)  # False negatives
    tn = np.sum(np.add(membership_data, detection_data.toarray()) == 0, axis=0).astype(np.uint)  # True negatives

    # Calculate Odds Ratio (OR)
    OR = ((tp + eps) / (fp + eps)) / ((fn + eps) / (tn + eps))  # Avoid division by zero using `eps`
    # OR = np.nan_to_num(OR)  # Handle NaN values explicitly if uncommented

    # Calculate Fold Enrichment (FE)
    FE = ((tp + eps) / (tp + fn + eps)) / ((tp + fp + eps) / (tp + fp + fn + tn + eps))
    
    # Fisher's exact test
    _, _, twosided = pvalue_npy(tp, fp, fn, tn)  # Two-sided p-value

    # Correct for multiple hypothesis testing
    _, p_adjusted, _, _ = multipletests(twosided, alpha=0.05, method=correction)

    # Compile results into a DataFrame
    df = pd.DataFrame({
        "Cluster": cluster,
        "TP": tp,
        "FP": fp,
        "FN": fn,
        "TN": tn,
        "FE": FE,
        "OR": OR,
        "logOR": np.log2(OR),
        "P": twosided,
        "P_adjust": p_adjusted,
        "log10P_adjust": -np.log10(p_adjusted)
    }, index=adata.var_names)

    # Filter results by significance, OR, and FE thresholds
    if return_significant:
        df = df[df['P_adjust'] <= p_threshold]
    if OR_cutoff is not None:
        df = df[df['OR'] >= OR_cutoff]
    if FE_cutoff is not None:
        df = df[df['FE'] >= FE_cutoff]

    # Clean up memory
    del detection_data
    del membership_data

    # Return sorted DataFrame by logOR
    return df.sort_values(by=['logOR'], ascending=[False])
```

```python
from joblib import Parallel, delayed  # Parallel processing utilities
import scanpy as sc  # Single-cell analysis library
import pandas as pd

def enrichment(adata, clusters, n_jobs=3):
    """
    Perform enrichment analysis by combining Odds Ratio and t-test results for single-cell data.
    
    Parameters:
    - adata: AnnData object with single-cell data
    - clusters: Column in `adata.obs` denoting clusters (must be categorical)
    - n_jobs: Number of parallel jobs to run
    
    Returns:
    - Dictionary containing enriched genes for each cluster with combined rankings and significance
    """
    # Validate that the cluster column is categorical
    if not pd.api.types.is_categorical_dtype(adata.obs[clusters]):
        raise ValueError(f"The column '{clusters}' must be categorical.")

    # Define a function to compute OR for a single cluster
    def compute_or(cluster):
        return OR(adata, cluster_col=clusters, cluster=cluster, eps=1e-7, correction='fdr_bh', return_significant=False)

    # Compute Odds Ratios in parallel for all clusters
    OR_results = Parallel(n_jobs=n_jobs, prefer="threads")(
        delayed(compute_or)(cluster) for cluster in adata.obs[clusters].cat.categories
    )
    
    # Convert OR results to a dictionary indexed by cluster
    OR_dict = {result['Cluster'].iloc[0]: result for result in OR_results}

    # Perform t-test for differential expression
    sc.tl.rank_genes_groups(
        adata, groupby=clusters, use_raw=False, method='t-test_overestim_var', n_genes=None, pts=True
    )

    # Combine Odds Ratio and t-test results
    DG_dict = {}
    for cluster in OR_dict.keys():
        # Extract Odds Ratio results and rank by OR
        OR_df = OR_dict[cluster].copy()
        OR_df['OR_rank'] = range(1, OR_df.shape[0] + 1)

        # Extract t-test results and rank by scores
        DEGs = sc.get.rank_genes_groups_df(adata, group=cluster).sort_values(by='scores', ascending=False)
        DEGs['lfc_rank'] = range(1, DEGs.shape[0] + 1)
        DEGs.set_index('names', inplace=True)  # Ensure gene names are used as the index

        # Merge results on gene names and calculate combined metrics
        DGs = pd.concat([OR_df, DEGs], axis=1, join='inner')
        DGs['combined_rank'] = DGs['OR_rank'] + DGs['lfc_rank']
        DGs['both_sig'] = (DGs['P_adjust'] < 0.05) & (DGs['pvals_adj'] < 0.05)
        DGs['either_sig'] = (DGs['P_adjust'] < 0.05) | (DGs['pvals_adj'] < 0.05)

        # Add to results dictionary
        DG_dict[cluster] = DGs

    return DG_dict
```

### Count number of cells in WT donor vs host samples

```python
# Count numbers of cells for each tissue in each sample type
import scipy.stats as st

host_cell_data = sdata.obs
host_cell_data = host_cell_data[host_cell_data.Condition == 'Host'][['Sample', 'Tissue_KO_clusters']].value_counts(sort=False).reset_index()
host_cell_data = host_cell_data.pivot(index='Tissue_KO_clusters', columns='Sample', values='count')
host_cell_data[host_cell_data.isna()] = 0

host_cell_data_pct = host_cell_data.div(host_cell_data.sum(axis=0), axis=1).multiply(100)
host_cell_data_pct['Host_Mean'] = host_cell_data_pct.mean(axis=1)
host_cell_data_pct = host_cell_data_pct.merge(pd.DataFrame(host_cell_data_pct.apply(lambda x: st.t.interval(0.95, len(x)-1, loc=np.mean(x), scale=st.sem(x)), axis=1).to_list(), index=host_cell_data_pct.index, columns=['CI95L', 'CI95U']), how='left', left_index=True, right_index=True)
host_cell_data_pct
```

```python
# Count numbers of cells for each tissue in each sample type
WT_cell_data = sdata.obs
WT_cell_data = WT_cell_data[WT_cell_data.Condition == 'WT'][['Sample', 'Tissue_KO_clusters']].value_counts(sort=False).reset_index()
WT_cell_data = WT_cell_data.pivot(index='Tissue_KO_clusters', columns='Sample', values='count')
WT_cell_data[WT_cell_data.isna()] = 0
WT_cell_data.columns = ['Donor_WT']
WT_cell_data_pct = WT_cell_data.div(WT_cell_data.sum(axis=0), axis=1).multiply(100)
total_cell_pct = host_cell_data_pct.merge(WT_cell_data_pct, how='left', left_index=True, right_index=True)
```

```python
# Concatenate results
total_cell_pct.loc[total_cell_pct['Donor_WT'].isna(), 'Donor_WT'] = 0
total_cell_pct['Donor_KO'] = 0
total_cell_pct['Below_range_WT'] = (total_cell_pct['Donor_WT'] < total_cell_pct['CI95L'])
total_cell_pct['Above_range_WT'] = (total_cell_pct['Donor_WT'] > total_cell_pct['CI95U'])
total_cell_pct['Outside_range_WT'] = total_cell_pct['Below_range_WT'] | total_cell_pct['Above_range_WT']
total_cell_pct['Outside_range_KO'] = (total_cell_pct['Donor_KO'] < total_cell_pct['CI95L']) | (total_cell_pct['Donor_KO'] > total_cell_pct['CI95U'])
# total_cell_pct.to_csv(f'{syncdir}/Cell_percentage_host_WT_confidence_interval.csv')
total_cell_pct
```

```python
import gseapy as gp
import pandas as pd

# Initialize dictionary to store enrichment results and a list for significant tissue terms
res_dict = {}
tissue_list = []

# Iterate through the defined KO clusters
for clust in ['KO 1', 'KO 2', 'KO 3', 'KO 4', 'KO 5']:
    # Retrieve top 500 genes for the current cluster
    gene_list = GOIs_all[clust][:500]  # GOIs_all should already have the top genes for each cluster
    print(f'Number of genes in {clust}: {len(gene_list)}')

    # Perform enrichment analysis using gseapy
    enr2 = gp.enrich(
        gene_list=gene_list,              # Input list of genes for the current cluster
        gene_sets=tissue_markers,         # Gene sets for enrichment (tissue-specific markers in this case)
        background=sdata.var.index.to_list(),  # Background genes (all genes in the dataset)
        cutoff=0.05,                      # P-value cutoff for significant enrichment
        outdir=None,                      # Do not save enrichment results to a local directory
        verbose=True                      # Verbose output for tracking progress
    )
    
    # Process and sort enrichment results
    results = enr2.results
    results = results.sort_values(
        ['Adjusted P-value', 'Odds Ratio', 'Term'],  # Sort by adjusted p-value, then odds ratio, then term
        ascending=[True, False, False]
    )
    results['log10_p'] = -np.log10(results['Adjusted P-value'])  # Convert adjusted p-value to log10 scale
    results['sig'] = results['Adjusted P-value'] <= 0.01        # Mark significant results with p <= 0.01

    # Save enrichment results for the current cluster to CSV files
    # results.to_csv(f'{syncdir}/Tissue_KO_clusters/Tissue_Marker_Enrichment_Tissue_KO_clusters_{clust}.csv')
    # results.to_csv(f'{outdir}/Tissue_KO_clusters/Tissue_Marker_Enrichment_Tissue_KO_clusters_{clust}.csv')

    # Print the results for tracking
    print(results)

    # Store results in the results dictionary
    res_dict[clust] = results

    # Append significant tissue marker terms to the tissue list
    tissue_list = tissue_list + results[results['sig']].Term.to_list()
```

```python
# Define receptor pathways and associated gene lists
RECs = {}
RECs['activin'] =  ['acvr1ba', 'acvr1bb', 'acvr1l', 'acvr2aa', 'acvr2ab', 'acvr2ba', 'acvr2bb', 'acvrl1']
RECs['bmp'] =  ['bmpr1aa', 'bmpr1ab', 'bmpr1ba', 'bmpr1bb', 'bmpr2a', 'bmpr2b']
RECs['tgfb'] =  ['tgfbr1a', 'tgfbr1b', 'tgfbr2a', 'tgfbr2b', 'tgfbr3']
RECs['fgf'] =  ['fgfr1a', 'fgfr1b', 'fgfr2', 'fgfr3', 'fgfr4', 'fgfrl1a', 'fgfrl1b']
RECs['fzd'] =  ['fzd1', 'fzd10', 'fzd2', 'fzd3a', 'fzd3b', 'fzd4', 'fzd5', 'fzd6', 'fzd7a', 'fzd7b', 'fzd8a', 'fzd8b', 'fzd9b']
RECs['hh'] = ['smo', 'ptch1', 'ptch2']
RECs['RA'] = ['raraa', 'rarab', 'rarga', 'rargb', 'rxraa', 'rxrba', 'rxrbb', 'rxrga', 'rxrgb']

# Optionally define additional pathways here, but commented for now:
# RECs['notch'] = ['notch1a', 'notch1b', 'notch2', 'notch3', 'notchl']
# RECs['gfr'] =  ['egfra', 'gfra1a', 'gfra1b', 'gfra3', 'gfra4a', 'ngfra', 'ngfrb', 'pdgfra', 'pdgfrb']
# RECs['cam'] = ['alcama', 'alcamb', 'bcam', 'cercam', 'epcam', 'mcama', 'mcamb', 'nrcama']
# RECs['cdh'] =  ['cdh1', 'cdh2', 'cdh4', 'cdh5', 'cdh6', 'cdh7a', 'cdh11', 'cdh13', 'cdh15', 'cdh17', 'cdh18a', 'cdh26.1', 'cdh29']
# RECs['eph'] =  ['epha2a', 'epha2b', 'epha3', 'epha4a', 'epha4b', 'epha4l', 'epha6', 'epha7', 'ephb1', 'ephb2a', 'ephb2b', 'ephb3a', 'ephb4a', 'ephb4b', 'ephb6']

# Process each receptor pathway
for clust in list(RECs.keys()):
    # Compute mean expression for receptor genes in each KO cluster
    X = sdata.to_df()[RECs[clust]].groupby(sdata.obs.KO_clusters).mean()
    
    # Normalize values within each pathway using MinMaxScaler
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    X[X.columns.to_list()] = scaler.fit_transform(X)
    
    # Transpose the DataFrame for clustering
    X = X.T
    names = X.index  # Gene names (used as labels)

    # Perform hierarchical clustering on the receptor pathway
    Z = scipy.cluster.hierarchy.linkage(X.values, method='complete')  # Complete linkage method
    dn = scipy.cluster.hierarchy.dendrogram(Z, labels=names.tolist(), orientation='right', no_plot=True)
    
    # Update the pathway order with the dendrogram leaf order
    RECs[clust] = dn['ivl']  # `ivl` contains the ordered labels after clustering

# Generate a dot plot to visualize the receptor pathways
sc.pl.dotplot(
    sdata, 
    RECs,  # Updated receptor pathway gene lists
    'Tissue_KO_clusters',  # Grouping variable in `obs`
    cmap='Reds',              # Colormap for dot intensity
    dendrogram=False,         # Disable dendrogram clustering in the plot
    standard_scale='var',     # Normalize values across genes
    dot_max=0.5,              # Cap the dot size to improve visualization
    # save=f'KO_cluster_receptor_PWs_expression.pdf'  # Save the plot as a PDF
)
```

### Visualise apoptosis data

```python
# Import the gseapy library for gene set enrichment analysis
import gseapy as gp

# Uncomment and use if you need to list available libraries for zebrafish
# zfish = gp.get_library_name(organism='Zebrafish')

# Fetch the GO biological process gene library for Zebrafish
go_bp = gp.get_library(name='GO_Biological_Process_2018', organism='Zebrafish')
term_names = list(go_bp.keys())  # Extract term names (GO categories)

# Filter terms containing 'apop' (apoptosis-related)
apop_terms = [i for i in term_names if 'apop' in i]
print(apop_terms[1])  # Debugging output: print the second term in the list

# Get genes associated with apoptosis-related GO terms
apoptotic_process1 = go_bp[apop_terms[0]]  # First apoptosis-related term
apoptotic_process2 = go_bp[apop_terms[1]]  # Second apoptosis-related term

# Combine and deduplicate gene lists from both terms
apoptotic_process = list(set(apoptotic_process1 + apoptotic_process2))

# Remove specific genes deemed uninformative
uninformative_genes = ['rpl11', 'rps3', 'tp53', 'zgc:110239', 'zgc:113149', 'zgc:153990', 
                       'zgc:153993', 'zgc:171731', 'zgc:194469']
apoptotic_process = [i for i in apoptotic_process if i not in uninformative_genes]

# Find genes of interest (GOIs) in single-cell data (sdata)
# Select genes matching apoptotic process symbols and convert to list
GOIs = sdata.var[sdata.var.gene_symbol.isin(apoptotic_process)].index.to_list()

# Sort GOIs for consistency
GOIs.sort()
```

```python
from scipy.stats import zscore  # Import z-score normalization function

# Create a DataFrame of apoptosis-related gene expression
# Extracts expression data for selected genes (GOIs) from the 'X_magic' layer
marker_df = pd.DataFrame(
    sdata.layers['X_magic'][:, sdata.var.index.isin(GOIs)].toarray(), 
    columns=sdata.var.index[sdata.var.index.isin(GOIs)], 
    index=sdata.obs.index
)

# Normalize expression data per gene using z-score scaling
marker_df = marker_df.apply(zscore)

# Cap values to remove influence of outliers and restrict range
# Set values below mean (< 0) to 0 and cap values above 3 SD to 3
marker_df[marker_df < 0] = 0
marker_df[marker_df > 3] = 3

# Compute apoptosis score: Sum normalized scores per cell and adjust by the number of genes and the cap
sdata.obs['Apoptosis_score'] = marker_df.sum(axis=1) / len(GOIs) / 3
```

```python
# Extract relevant columns from sdata.obs for apoptosis scoring
data = sdata.obs[['Tissue_KO_clusters', 'Apoptosis_score']]

# Group data by Tissue_KO_clusters and calculate mean apoptosis score per group
apoptosis_mean_order = (
    data.groupby('Tissue_KO_clusters')
    .mean()
    .sort_values('Apoptosis_score', ascending=False)
)

# Add cluster names as a column for visualization
apoptosis_mean_order['Tissue_KO_clusters'] = apoptosis_mean_order.index.to_list()

# Map cluster names to their corresponding colors using pre-defined categorical color mapping
lut = dict(zip(sdata.obs['Tissue_KO_clusters'].cat.categories, sdata.uns['Tissue_KO_clusters_colors']))
apoptosis_mean_order['Colors'] = apoptosis_mean_order['Tissue_KO_clusters'].map(lut)

# Display the ordered DataFrame with apoptosis scores and colors
apoptosis_mean_order
```

```python
# Reorder the categories of 'Tissue_KO_clusters' in sdata.obs based on the apoptosis score order
sdata.obs['Tissue_KO_clusters'] = sdata.obs['Tissue_KO_clusters'].cat.reorder_categories(
    apoptosis_mean_order.index.to_list()
)

# Update the cluster colors in sdata.uns with the reordered color list from apoptosis_mean_order
sdata.uns['Tissue_KO_clusters_colors'] = apoptosis_mean_order.Colors.to_list()

# Plot a violin plot of Apoptosis_score by Tissue_KO_clusters
sc.pl.violin(
    sdata,  # AnnData object containing the data
    'Apoptosis_score',  # Variable to plot
    'Tissue_KO_clusters',  # Grouping variable
    rotation=90,  # Rotate x-axis labels for better readability
    use_raw=False,  # Use processed data instead of raw data
    Title='Imputed apoptosis geneset score',
    # save=f'.QC_{study}_Apoptosis_score_violin_Tissue_KO_clusters.pdf'  # Save plot to file with descriptive name
)
```

```python
# Create a heatmap of imputed apoptosis gene expression

# Create a DataFrame of expression data from the 'X_magic' layer
exp_data = pd.DataFrame(
    sdata.layers['X_magic'].toarray(), 
    columns=sdata.var.index, 
    index=sdata.obs.index
)

# Select only genes of interest (GOIs)
exp_data = exp_data.loc[:, GOIs]

# Add Tissue_KO_clusters as a categorical column and reorder categories
exp_data['Tissue_KO_clusters'] = pd.Categorical(
    sdata.obs['Tissue_KO_clusters'], 
    categories=apoptosis_mean_order.index.to_list(), 
    ordered=True
)

# Sort expression data by Tissue_KO_clusters
exp_data = exp_data.sort_values('Tissue_KO_clusters')

# Keep the top rows per cluster for visualization (parameterized)
TOP_ROWS = 50  # Define the number of rows to retain per cluster
exp_data = exp_data.groupby('Tissue_KO_clusters').head(TOP_ROWS)

# Extract cluster information for row coloring
col_data = exp_data[['Tissue_KO_clusters']]
exp_data = exp_data.drop(columns=['Tissue_KO_clusters'])  # Remove cluster column from expression data

# Map clusters to their corresponding colors
lut = dict(zip(sdata.obs['Tissue_KO_clusters'].cat.categories, sdata.uns['Tissue_KO_clusters_colors']))

# Validate color mapping to ensure no missing clusters
missing_clusters = set(col_data['Tissue_KO_clusters']).difference(lut.keys())
if missing_clusters:
    raise ValueError(f"Missing colors for clusters: {missing_clusters}")

# Map row colors for heatmap
row_colors = col_data['Tissue_KO_clusters'].map(lut)

# Parameterize clustering and colormap
CLUSTERING_METHOD = 'ward'
HEATMAP_CMAP = 'magma'

# Create a clustered heatmap of apoptosis gene expression
sns.clustermap(
    exp_data, 
    standard_scale=1,  # Normalize each gene to mean 0 and variance 1
    method=CLUSTERING_METHOD,  # Clustering method
    cmap=HEATMAP_CMAP,  # Colormap for heatmap
    xticklabels=True,  # Show gene names on the x-axis
    yticklabels=False,  # Hide row labels
    row_cluster=False,  # Do not cluster rows (maintain cluster sorting)
    row_colors=row_colors  # Add row colors for clusters
)

# Save the plot to a file (uncomment to activate saving)
# plt.savefig(f'{syncdir}/Apoptosis_gene_heatmap_Tissue_KO_clusters.pdf')

# Display the plot
plt.show()
```
