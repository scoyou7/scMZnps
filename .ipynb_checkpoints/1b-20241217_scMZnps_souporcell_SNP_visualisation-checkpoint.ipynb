### Output list of barcodes associated with KO, WT and Host cells

```python
import io
import os
import pandas as pd
import gzip

def read_vcf(path):
    """
    Reads a gzip-compressed VCF file and loads it into a Pandas DataFrame.

    Parameters:
        path (str): Path to the gzip-compressed VCF file.

    Returns:
        pd.DataFrame: A DataFrame containing the VCF data, with the header row
                      as column names and the data rows as table content.

    Raises:
        FileNotFoundError: If the provided VCF file path does not exist.
        ValueError: If there is an issue reading or parsing the VCF file.
    """
    # Validate the input path
    if not os.path.exists(path):
        raise FileNotFoundError(f"File not found: {path}")
    
    try:
        # Open the gzip-compressed VCF file
        with gzip.open(path, 'rt') as f:
            # Read lines, skipping metadata lines (those starting with "##")
            lines = [l for l in f if not l.startswith('##')]
        
        # Combine the remaining lines and parse into a Pandas DataFrame
        return pd.read_csv(
            io.StringIO(''.join(lines)),  # Treat the combined string as a file-like object
            dtype={
                '#CHROM': str,  # Chromosome identifier
                'POS': int,     # Position on the chromosome
                'ID': str,      # Variant identifier
                'REF': str,     # Reference base(s)
                'ALT': str,     # Alternate base(s)
                'QUAL': str,    # Quality score
                'FILTER': str,  # Filter status
                'INFO': str     # Additional information
            },
            sep='\t'  # Tab-separated fields in VCF
        ).rename(columns={'#CHROM': 'CHROM'})  # Rename '#CHROM' to 'CHROM' for convenience
    
    except Exception as e:
        # Handle exceptions raised during file reading or parsing
        raise ValueError(f"Error reading VCF file at {path}: {e}")
```

### Process mutant A sample

```python
import scipy.io

# Step 1: Load alternative allele count matrix
# The alt.mtx file contains the alternative allele counts for each variant
sdata_alt_features_1 = sc.read_mtx(
    '/plus/scratch/@scratch_scott/projects/scmznos_valtor/project_results/souporcell/MUT_CD4_chr1/alt.mtx'
).transpose()
# Transpose the matrix to align variants as columns (genes/features) and cells as rows

# Step 2: Load reference allele count matrix and store in layers
# Read the reference allele count matrix (ref.mtx)
sdata_alt_features_1.layers['X_ref'] = scipy.io.mmread(
    '/plus/scratch/@scratch_scott/projects/scmznos_valtor/project_results/souporcell/MUT_CD4_chr1/ref.mtx'
).transpose().tocsr()
# Transpose and convert to CSR format for efficient storage and operations

# Step 3: Load variant information from VCF file
# Use the read_vcf function to load the VCF file into a Pandas DataFrame
sdata_alt_features_1.var = read_vcf(
    '/plus/scratch/@scratch_scott/projects/scmznos_valtor/project_results/souporcell/MUT_CD4_chr1/souporcell_merged_sorted_vcf.vcf.gz'
)

# Step 4: Create unique identifiers for each variant
# Combine chromosome, position, reference allele, and alternate allele into a single index
sdata_alt_features_1.var.index = (
    'chr' +
    sdata_alt_features_1.var.CHROM.astype('str') + ":" +
    sdata_alt_features_1.var.POS.astype('str') + "-" +
    sdata_alt_features_1.var.REF + sdata_alt_features_1.var.ALT
)
```

```python
# Step 1: Import cell metadata
# Read the clusters.tsv file containing barcode and cluster assignments
cells = pd.read_csv(
    '/plus/scratch/@scratch_scott/projects/scmznos_valtor/project_results/souporcell/MUT_CD4_chr1/clusters.tsv',
    sep='\t'  # Tab-separated file
)
cells.index = cells.barcode  # Set the barcode column as the index for easy alignment

# Step 2: Assign cell metadata to AnnData object
# Add the imported cell metadata to the AnnData object (sdata_alt_features_1)
sdata_alt_features_1.obs = cells

# Step 3: Subset the AnnData object to include only QC-filtered cells
# Filter the AnnData object to include only cells present in sdata.obs
sdata_alt_features_1 = sdata_alt_features_1[sdata_alt_features_1.obs.index.isin(sdata.obs.index)]

# Step 4: Subset the imported cell metadata to match the filtered AnnData object
# Filter the cells DataFrame to include only cells present in the filtered AnnData object
cells = cells[cells.index.isin(sdata_alt_features_1.obs.index)]

# Step 5: Merge metadata
# Concatenate the filtered cells metadata with metadata from sdata.obs for 'Mutant_A' samples
# The axis=1 option combines columns based on their indices
sdata_alt_features_1.obs = pd.concat(
    [cells, sdata.obs[sdata.obs.Sample == 'Mutant_A']], 
    axis=1
)

# Inspect the updated AnnData object
sdata_alt_features_1
```

```python
# Map values in the 'assignment' column to ensure consistency with Mutant B variant calling
mapping = {'1': '0', '0': '1'}
sdata_alt_features_1.obs['assignment'] = sdata_alt_features_1.obs['assignment'].map(mapping)
```

```python
# Validate assignment column and expected values
if 'assignment' not in sdata_alt_features_1.obs.columns:
    raise ValueError("'assignment' column not found in AnnData object.")
if not {'1', '0'}.issubset(sdata_alt_features_1.obs['assignment'].unique()):
    raise ValueError("'assignment' column does not contain expected values ('1' for KO and '0' for WT).")

# Helper function to calculate and assign QC metrics
def calculate_qc_metrics_by_group(sdata, group_label):
    _, stats = sc.pp.calculate_qc_metrics(sdata[sdata.obs.assignment == group_label], inplace=False)
    return {
        'n_cells_by_counts': stats.n_cells_by_counts,
        'undetected': stats.n_cells_by_counts == 0,
        'total_counts': stats.total_counts,
        'mean_counts': stats.mean_counts,
        'mean_across_detected': stats.total_counts / stats.n_cells_by_counts.replace(0, pd.NA)
    }

# Step 1: Calculate QC metrics for KO cells
ko_metrics = calculate_qc_metrics_by_group(sdata_alt_features_1, '1')
for key, value in ko_metrics.items():
    sdata_alt_features_1.var[f'KO_{key}'] = value

# Step 2: Calculate QC metrics for WT cells
wt_metrics = calculate_qc_metrics_by_group(sdata_alt_features_1, '0')
for key, value in wt_metrics.items():
    sdata_alt_features_1.var[f'WT_{key}'] = value

# Step 3: Compute differences between WT and KO
sdata_alt_features_1.var['Delta_WT_KO_n_cells_by_counts'] = (
    sdata_alt_features_1.var['KO_n_cells_by_counts'] - sdata_alt_features_1.var['WT_n_cells_by_counts']
)

# Log results
print("Calculated QC metrics for KO and WT.")
```

```python
# Get top SNPS only found in KO cells
VOIs = sdata_alt_features_1.var[
    (~sdata_alt_features_1.var['KO_undetected']) & 
    (sdata_alt_features_1.var['WT_undetected'])
]['Delta_WT_KO_n_cells_by_counts'].sort_values(ascending=False).head(10).index
VOIs
```

```python
# Subset sdata and assign PHATE embedding
subset = sdata[sdata.obs.Sample.str.contains('Mutant_A', na=False)]
sdata_alt_features_1.obsm['X_phate'] = subset.obsm['X_phate']

# Ensure 'assignment' is categorical
sdata_alt_features_1.obs['assignment'] = sdata_alt_features_1.obs['assignment'].astype('category')

# Adjust number of columns for visualization
ncols = min(len(VOIs), 5)

# Plot embedding
sc.pl.embedding(
    sdata_alt_features_1,
    basis='X_phate',
    color=VOIs,
    cmap='Reds',
    vmin=0,
    vmax=1,
    add_outline=False,
    legend_fontsize=12,
    legend_fontoutline=2,
    frameon=False,
    ncols=ncols
)
```

### Find common features between the 2 samples

```python
# Find common features between datasets
common_features = sdata_alt_features_1.var.index.intersection(sdata_alt_features_2.var.index)

# Add sample identifiers to differentiate metadata
sdata_alt_features_1.obs['Sample'] = 'Sample_1'
sdata_alt_features_2.obs['Sample'] = 'Sample_2'

# Subset and concatenate datasets
sdata_alt_features = sc.concat(
    [sdata_alt_features_1[:, common_features], sdata_alt_features_2[:, common_features]],
    join='inner'  # Ensure that metadata aligns properly
)

# Inspect merged dataset
sdata_alt_features
```

```python
# Define function for calculating metrics
def calculate_metrics(data, label):
    _, stats = sc.pp.calculate_qc_metrics(data[data.obs.assignment == label], inplace=False)
    return {
        'n_cells_by_counts': stats.n_cells_by_counts,
        'undetected': stats.n_cells_by_counts == 0,
        'total_counts': stats.total_counts,
        'mean_counts': stats.mean_counts,
        'mean_across_detected': stats.total_counts / stats.n_cells_by_counts.replace(0, pd.NA)
    }

# Calculate KO and WT metrics
ko_metrics = calculate_metrics(sdata_alt_features, '1')
wt_metrics = calculate_metrics(sdata_alt_features, '0')

# Assign metrics to AnnData object
for key, value in ko_metrics.items():
    sdata_alt_features.var[f'KO_{key}'] = value

for key, value in wt_metrics.items():
    sdata_alt_features.var[f'WT_{key}'] = value

# Calculate differences
sdata_alt_features.var['Delta_WT_KO_n_cells_by_counts'] = (
    sdata_alt_features.var['KO_n_cells_by_counts'] - sdata_alt_features.var['WT_n_cells_by_counts']
)

# Inspect results
print("Metrics calculated for KO and WT cells.")
```

```python
# Parse Variants of Interest (VOIs) into a DataFrame
VOI_df = pd.DataFrame(
    VOIs.index.str.split(':|-', expand=True).to_list(), 
    columns=['Chromosome', 'Base', 'Mutation'], 
    index=VOIs.index
)
VOI_df[['Ref', 'Alt']] = VOI_df.Mutation.str.split('', expand=True).iloc[:, 1:3]
VOI_df['Base'] = VOI_df['Base'].astype(int)
VOI_df
```

```python
# Bar plot of SNP frequency in bulk genome samples (known genotype)

import pysam
import os
import matplotlib.pyplot as plt

# Define BAM file paths
bamKO = '/home/scott/public/mznps_AGR002963_sorted.bam'
bamWT = '/home/scott/public/wt_AGR002965_sorted.bam'

# Validate BAM file paths
for bam_file in [bamKO, bamWT]:
    if not os.path.exists(bam_file):
        raise FileNotFoundError(f"BAM file not found: {bam_file}")

# Initialize result lists
WT_ref, WT_alt, WT_rc = [], [], []
KO_ref, KO_alt, KO_rc = [], [], []

for i in VOI_df.index.to_list():
    # Get base of interest (0-indexed)
    BOI = VOI_df.Base[i] - 1

    # Process KO BAM file
    KO_res = []
    with pysam.AlignmentFile(bamKO, "rb") as KOsamfile:
        for pileupcolumn in KOsamfile.pileup('1', BOI, BOI + 1):
            for pileupread in pileupcolumn.pileups:
                if pileupread.query_position is None:
                    continue
                if pileupcolumn.pos == BOI:
                    KObase = pileupread.alignment.query_sequence[pileupread.query_position]
                    KO_res.append(KObase)

    KO_reads = len(KO_res)
    KO_rc.append(KO_reads)

    if KO_reads > 0:
        KO_ref.append(sum([j == VOI_df.Ref[i] for j in KO_res]) / KO_reads * 100)
        KO_alt.append(sum([j == VOI_df.Alt[i] for j in KO_res]) / KO_reads * 100)
    else:
        KO_ref.append(0)
        KO_alt.append(0)

    # Process WT BAM file
    WT_res = []
    with pysam.AlignmentFile(bamWT, "rb") as WTsamfile:
        for pileupcolumn in WTsamfile.pileup('1', BOI, BOI + 1):
            for pileupread in pileupcolumn.pileups:
                if pileupread.query_position is None:
                    continue
                if pileupcolumn.pos == BOI:
                    WTbase = pileupread.alignment.query_sequence[pileupread.query_position]
                    WT_res.append(WTbase)

    WT_reads = len(WT_res)
    WT_rc.append(WT_reads)

    if WT_reads > 0:
        WT_ref.append(sum([j == VOI_df.Ref[i] for j in WT_res]) / WT_reads * 100)
        WT_alt.append(sum([j == VOI_df.Alt[i] for j in WT_res]) / WT_reads * 100)
    else:
        WT_ref.append(0)
        WT_alt.append(0)

# Add results to VOI_df
VOI_df['KO_counts'], VOI_df['KO_ref'], VOI_df['KO_alt'] = KO_rc, KO_ref, KO_alt
VOI_df['WT_counts'], VOI_df['WT_ref'], VOI_df['WT_alt'] = WT_rc, WT_ref, WT_alt

# Plot SNP frequency for the first variant
variant = VOI_df.index.to_list()[0]
ax = VOI_df.loc[variant, ['KO_alt', 'WT_alt']].plot.bar(title=f"SNP Frequencies: {variant}")
ax.set_xlabel("Condition")
ax.set_ylabel("Percentage")
plt.tight_layout()
plt.savefig(f"{syncdir}/SNP_frequency_{variant.replace(':', '_')}.pdf")
plt.show()
```

```python
# Bar plot of SNP frequency in bulk genome samples (known genotype)

import pysam
import os
import matplotlib.pyplot as plt

# Define BAM file paths
bamKO = '/home/scott/public/mznps_AGR002963_sorted.bam'
bamWT = '/home/scott/public/wt_AGR002965_sorted.bam'

# Validate BAM file paths
for bam_file in [bamKO, bamWT]:
    if not os.path.exists(bam_file):
        raise FileNotFoundError(f"BAM file not found: {bam_file}")

# Initialize result lists
WT_ref, WT_alt, WT_rc = [], [], []
KO_ref, KO_alt, KO_rc = [], [], []

for i in VOI_df.index.to_list():
    # Get base of interest (0-indexed)
    BOI = VOI_df.Base[i] - 1

    # Process KO BAM file
    KO_res = []
    with pysam.AlignmentFile(bamKO, "rb") as KOsamfile:
        for pileupcolumn in KOsamfile.pileup('1', BOI, BOI + 1):
            for pileupread in pileupcolumn.pileups:
                if pileupread.query_position is None:
                    continue
                if pileupcolumn.pos == BOI:
                    KObase = pileupread.alignment.query_sequence[pileupread.query_position]
                    KO_res.append(KObase)

    KO_reads = len(KO_res)
    KO_rc.append(KO_reads)

    if KO_reads > 0:
        KO_ref.append(sum([j == VOI_df.Ref[i] for j in KO_res]) / KO_reads * 100)
        KO_alt.append(sum([j == VOI_df.Alt[i] for j in KO_res]) / KO_reads * 100)
    else:
        KO_ref.append(0)
        KO_alt.append(0)

    # Process WT BAM file
    WT_res = []
    with pysam.AlignmentFile(bamWT, "rb") as WTsamfile:
        for pileupcolumn in WTsamfile.pileup('1', BOI, BOI + 1):
            for pileupread in pileupcolumn.pileups:
                if pileupread.query_position is None:
                    continue
                if pileupcolumn.pos == BOI:
                    WTbase = pileupread.alignment.query_sequence[pileupread.query_position]
                    WT_res.append(WTbase)

    WT_reads = len(WT_res)
    WT_rc.append(WT_reads)

    if WT_reads > 0:
        WT_ref.append(sum([j == VOI_df.Ref[i] for j in WT_res]) / WT_reads * 100)
        WT_alt.append(sum([j == VOI_df.Alt[i] for j in WT_res]) / WT_reads * 100)
    else:
        WT_ref.append(0)
        WT_alt.append(0)

# Add results to VOI_df
VOI_df['KO_counts'], VOI_df['KO_ref'], VOI_df['KO_alt'] = KO_rc, KO_ref, KO_alt
VOI_df['WT_counts'], VOI_df['WT_ref'], VOI_df['WT_alt'] = WT_rc, WT_ref, WT_alt

# Plot SNP frequency for the first variant
variant = VOI_df.index.to_list()[0]
ax = VOI_df.loc[variant, ['KO_alt', 'WT_alt']].plot.bar(title=f"SNP Frequencies: {variant}")
ax.set_xlabel("Condition")
ax.set_ylabel("Percentage")
plt.tight_layout()
plt.savefig(f"{syncdir}/SNP_frequency_{variant.replace(':', '_')}.pdf")
plt.show()
```
